```python
# ‚úÖ Installer les biblioth√®ques n√©cessaires
!pip install -U fastai fastdownload duckduckgo-search gradio

# ‚úÖ Importer les biblioth√®ques
from fastai.vision.all import *
from fastdownload import download_url
from duckduckgo_search import DDGS
import gradio as gr
import time

# ‚úÖ Cr√©er le dossier projet
project_name = 'cats_or_dogs'
path = Path(project_name)
path.mkdir(exist_ok=True)

# ‚úÖ Fonction de recherche et t√©l√©chargement d'images
def search_images(term, max_images=20):
    print(f"üîç Recherche d'images pour : {term}")
    urls = []
    with DDGS() as ddgs:
        results = ddgs.images(term, max_results=max_images)
        for r in results:
            urls.append(r["image"])
    return urls

# ‚úÖ T√©l√©charger les images pour chaque classe
searches = 'cat','dog'
for o in searches:
    dest = (path/o)
    dest.mkdir(exist_ok=True, parents=True)
    urls = search_images(f'{o} photo', max_images=30)
    download_images(dest, urls=urls)
    time.sleep(5)

    # ‚úÖ Redimensionner sans bloquer sur AVIF
    for img_path in get_image_files(dest):
        if img_path.suffix.lower() in ['.jpg', '.jpeg', '.png']:
            try:
                img = PILImage.create(img_path)
                img.thumbnail((400, 400))
                img.save(img_path)
            except Exception as e:
                print(f"‚ùå Erreur {e} sur {img_path.name}, suppression")
                img_path.unlink()
        else:
            print(f"‚ùå Format non support√© {img_path.suffix}, suppression {img_path.name}")
            img_path.unlink()

# ‚úÖ Supprimer les images corrompues restantes
failed = verify_images(get_image_files(path))
failed.map(Path.unlink)
print(f"üóëÔ∏è Images corrompues supprim√©es : {len(failed)}")

# ‚úÖ Cr√©er les DataLoaders
dls = DataBlock(
    blocks=(ImageBlock, CategoryBlock),
    get_items=get_image_files,
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=Resize(192)
).dataloaders(path, bs=16)

# ‚úÖ Visualiser un batch pour valider
dls.show_batch(max_n=6)

# ‚úÖ Entra√Æner le mod√®le
learn = vision_learner(dls, resnet18, metrics=accuracy)
learn.fine_tune(3)

# ‚úÖ Exporter le mod√®le
learn.export('cats_vs_dogs.pkl')
print("‚úÖ Mod√®le export√© avec succ√®s !")

# ‚úÖ Charger le mod√®le pour Gradio
learn = load_learner('cats_vs_dogs.pkl')

# ‚úÖ Fonction pr√©diction Gradio
def predict_image(img):
    pred, idx, probs = learn.predict(img)
    return {pred: float(probs[idx])}

# ‚úÖ Lancer Gradio pour d√©mo avec lien partageable
gr.Interface(
    fn=predict_image,
    inputs=gr.Image(type="pil"),
    outputs=gr.Label(),
    title="üê±üê∂ Classification IA Chat ou Chien",
    description="Upload une image pour obtenir une pr√©diction instantan√©e gr√¢ce √† l'IA entra√Æn√©e."
).launch(share=True)
```
