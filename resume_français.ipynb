```python
!pip install -U transformers gradio

from transformers import pipeline
import gradio as gr

# ✅ Utiliser un modèle français
summarizer = pipeline("summarization", model="plguillou/t5-base-fr-sum-cnndm")

def summarize_text(text):
    # Reformulation synthétique pour emails
    prompt = f"Réécris de manière synthétique en français : {text}"
    summary = summarizer(prompt, max_length=80, min_length=20, do_sample=False)
    return summary[0]['summary_text']

gr.Interface(
    fn=summarize_text,
    inputs=gr.Textbox(lines=10, placeholder="Collez ici votre email à résumer..."),
    outputs="text",
    title="📩 Résumeur IA d'emails en français",
    description="Obtenez un résumé synthétique et clair de vos emails en français grâce à l'IA."
).launch(share=True)
```
